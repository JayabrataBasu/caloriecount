```python
print("Hello World!")
import pandas as pd

# Load the CSV file
file_path = 'data/gym_members_exercise_tracking.csv'  # Adjust path as needed
data = pd.read_csv(file_path)

# Display the first few rows to confirm it's loaded correctly
data.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Specify the path to the CSV file
# If the file is in a 'data' folder, use this path:
file_path = os.path.join("data", "gym_members_exercise_tracking.csv")

# Load the data
data = pd.read_csv(file_path)

# Set up the visual style for plots
sns.set(style="whitegrid")

# Calories burned by workout type
plt.figure(figsize=(12, 8))
sns.boxplot(x="Workout_Type", y="Calories_Burned", data=data, palette="Set2")
plt.title("Calories Burned by Workout Type")
plt.xlabel("Workout Type")
plt.ylabel("Calories Burned")
plt.xticks(rotation=45)
plt.show()
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
# Summary statistics
summary_stats = data[['Avg_BPM', 'BMI', 'Workout_Frequency (days/week)']].describe()
print(summary_stats)

plt.figure(figsize=(10, 6))
sns.boxplot(x='Workout_Type', y='Avg_BPM', data=data)
plt.title('Heart Rate Distribution by Exercise Type')
plt.xlabel('Exercise Type')
plt.ylabel('Heart Rate (bpm)')
plt.show()
```

```python
# Summary statistics
summary_stats = data[['Max_BPM', 'Avg_BPM', 'Resting_BPM', 'Calories_Burned', 'BMI']].describe()
print(summary_stats)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Workout_Type', y='Avg_BPM', data=data)
plt.title('Average BPM Distribution by Workout Type')
plt.xlabel('Workout Type')
plt.ylabel('Average BPM')
plt.xticks(rotation=45)
plt.show()
plt.figure(figsize=(10, 6))
sns.histplot(data['BMI'], bins=20, kde=True)
plt.title('Distribution of BMI')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Session_Duration (hours)', y='Calories_Burned', data=data)
plt.title('Calories Burned vs. Session Duration')
plt.xlabel('Session Duration (hours)')
plt.ylabel('Calories Burned')
plt.show()
# Correlation matrix
correlation_matrix = data[['Max_BPM', 'Avg_BPM', 'Resting_BPM', 'Calories_Burned', 'BMI']].corr()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
